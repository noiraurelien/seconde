```python
Z=int(input("Quelle est la valeur du numéro atomique ? "))   # on entre le numéro atomique de l'atome étudié, le résultat est converti en une variable de type entier
print("La configuration électronique de l'atome de numéro atomique",Z,"est:")
if Z <= 2 :       # condition de type if (si..)
    print("(1s",Z,")")
```
